## tokenization  converting paragraph into sentences and words

```python
nltk.download('punkt')
sentences = nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
stemmer = PorterStemmer()
```

```python
stemmer.stem('goes')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize('goes')
```

## Cleaning sentences with regular expressions


```python
len(sentences)
```

```python
import re
corpus = []
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ', sentences[i])
    review = review.lower()
    corpus.append(review)

```

```python
len(corpus)
```

### Stemming

```python
stopwords.words('english')
```

```python
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(stemmer.stem(word))
        
```

### lematize

```python
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(lemmatizer.lemmatize(word))
```

### Apply lemmatize and Stopwords

```python
import re
corpus = []
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ', sentences[i])
    review = review.lower()
    reviews = review.split()
    reviews = [lemmatizer.lemmatize(word) for word in reviews if not word in set(stopwords.words('english'))]
    review = ' '.join(reviews)
    corpus.append(review)
```

```python
corpus
```

```python
cv = CountVectorizer(binary=True, ngram_range=(2, 3))
```

```python
X = cv.fit_transform(corpus)
```

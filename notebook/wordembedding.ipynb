## Word Embedding Techniques using  Embedding Layer in Keras

```python
from  tensorflow.keras.preprocessing.text import one_hot
```

```python
import tensorflow as tf
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
sent
```

```python
### Vocabulary size
voc_size=10000
```

```python
## OneHot Representation
onehot_repr = [one_hot(words,voc_size) for words in sent]
```

```python
print(onehot_repr)
```

## Word Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Hashing
```

```python
## PREPADING SENTECE FOR WORD EMBEDDING
sent_length = 8
embeded_docs = pad_sequences(onehot_repr, padding='pre', maxlen=sent_length)
print(embeded_docs)
```

```python
## Coverting embedding to vector representation with feature dimension of 10
dim =10
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim, input_length=sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
embeded_docs[0]
```

```python
model.predict(embeded_docs[0])
```
